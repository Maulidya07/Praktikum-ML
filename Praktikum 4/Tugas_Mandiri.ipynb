# Membaca dataset

```python
from pathlib import Path
#memanggil data set lewat gdrive
Path = "/content/gdrive/MyDrive/Praktikum_ML/Praktikum4/data"
```

```python
import pandas as pd

df = pd.read_csv(Path + '/calonpembelimobil.csv')
df.head()
```

# Pisah menjadi fitur dan target

```python
X = df[['Usia', 'Status', 'Kelamin', 'Memiliki_Mobil', 'Penghasilan']]
y = df['Beli_Mobil']
```

#Mmembagi data menjadi trainset

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

# Buat dan Latih model

```python
# Buat dan latih model
model = LogisticRegression()
model.fit(X_train, y_train)
print("Model Logistic Regression berhasil dilatih!\n")
```

```python
# Prediksi data uji
y_pred = model.predict(X_test)

# Evaluasi model
print("Akurasi Model:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nLaporan Klasifikasi:\n", classification_report(y_test, y_pred))
```

# Uji dengan dataset baru

```python
data_baru = pd.DataFrame({
    'Usia': [31, 22, 19],
    'Status': [2, 1, 2],
    'Kelamin': [1, 1, 0],
    'Memiliki_Mobil': [5, 3, 2],
    'Penghasilan': [700, 435, 300]
})

# Prediksi dengan model
prediksi = model.predict(data_baru)

# Tambahkan hasil ke tabel
data_baru['Prediksi_Beli_Mobil'] = prediksi
print(data_baru)
```
