```python
from pathlib import Path
#memanggil data set lewat gdrive
Path = "/content/gdrive/MyDrive/Praktikum_ML/Praktikum2/data"
```

# Membaca File CSV

```python
import pandas as pd

df = pd.read_csv(Path + '/day.csv')
df.head()
```

# Jawaban A

```python
#Membagi dataset menjadi 80% data training dari total 100% dataset
from sklearn.model_selection import train_test_split

data_training, data_sisa = train_test_split(df, test_size=0.2, random_state=42)

# Menampilkan hasil
print("Jumlah Data:")
print(f"Total Dataset  : {len(df)}")
print(f"Data Training  : {len(data_training)}")
print(f"Data Sisa (20%): {len(data_sisa)}")

print("\n 5 Data Teratas dari Data Training:")
display(data_training.head())
```

# Jawaban B

```python
# Membagi 80% data training dan 20% sisa
data_training_awal, data_sisa = train_test_split(df, test_size=0.2, random_state=42)

# Sekarang bagi lagi 10% dari data training menjadi data validation
data_training, data_validation = train_test_split(data_training_awal, test_size=0.1, random_state=42)

# Tampilkan hasil pembagian
print("Jumlah Data:")
print(f"Data Training Awal (80%) : {len(data_training_awal)}")
print(f"Data Training (90%)      : {len(data_training)}")
print(f"Data Validation (10%)    : {len(data_validation)}")

print("\n5 Data Teratas dari Data Validation:")
display(data_validation.head())
```

# Jawaban C

```python
# Membagi 20% data testing dari total 100% dataset
_, data_testing = train_test_split(df, test_size=0.2, random_state=42)

# Tampilkan hasil
print("Jumlah Data Testing:", len(data_testing))
print("\n5 Data Teratas dari Data Testing:")
display(data_testing.head())
```
